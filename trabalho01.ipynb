### Parte 1: Utilize um software para as representações dos sinais no domínio do tempo e da frequência

Considere os seguintes sinais de mensagem em banda base:

- $v_m(t) = \cos(2 \cdot \pi \cdot 1000 \cdot t)$
- $v_m(t) = \cos{(2 \cdot \pi \cdot 1000 \cdot t)} + \cos{(2 \cdot \pi \cdot 2000 \cdot t)}$
- $v_m(t) = \cos(2 \cdot \pi \cdot 1000 \cdot t) \cdot \cos{(2 \cdot \pi \cdot 3000 \cdot t)}$

Considere uma portadora dada por: $v_c(t) = \cos{(2 \cdot \pi \cdot 10000 \cdot t)}$

```python
import matplotlib.pyplot as plt
import numpy as np

mi = 0.8
```

```python
function = lambda t : np.cos(2 * np.pi * 1000 * t)
carrier = lambda t : np.cos(2 * np.pi * 10000 * t)
samples = 128
t = np.linspace(-10, 10, samples)

fig, ax = plt.subplots(3, 2)

fig.tight_layout()

ax[0][0].set_title("cos(2π * 1000t)")
ax[0][0].plot(t, function(t), color="blue")

ax[0][1].set_title("Espectro Modulante")
ax[0][1].plot(t, np.abs(np.fft.fftfreq(samples, function(t))), color="green")

ax[1][0].set_title("Portadora")
ax[1][0].plot(t, carrier(t), color="blue")

ax[1][1].set_title("Espectro Portadora")
ax[1][1].plot(t, np.abs(np.fft.fftfreq(samples, carrier(t))), color="green")

ax[2][0].set_title("Modulada")
ax[2][0].plot(t, (1 + mi * function(t)) * carrier(t), color="blue")
ax[2][0].plot(t, (function(t) + 1), color="red")
ax[2][0].plot(t, (function(t) - 1), color="red")

ax[2][1].set_title("Espectro Modulada")
ax[2][1].plot(t, np.abs(np.fft.fftfreq(samples, (1 + mi * function(t)) * carrier(t))), color="green")

plt.show()
```
